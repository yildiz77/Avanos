```python
with open('data_file/sample.txt') as f:
    line = f.readline()
    
    while line:
        print(line)
        line = f.readline()
```

###   `write` a file <br>

#### * `file.write(string)` writes the contents of string to the file, returning the number of characters written. 
     


####  * We need to be careful with the 'w' mode as it will overwrite into the file if it already exists. All previous data are erased.
<br>

This program will create a new file named `sample.txt` if it does not exist. If it does exist, it is overwritten.

```python
file = open('data_file/sample.txt', 'w') 

file.close()
```

```python
!cat data_file/sample.txt
```

```python
file = open('data_file/sample.txt', 'w')    

file.write("Let's check the write operation")

file.close()
```

```python
!cat data_file/sample.txt
```

<br><br>

```python
file = open('data_file/sample.txt', 'w')  

file.write("Let's check the write operation")

file.seek(5)

file.write(" examine ")

file.close()
```

```python
!cat data_file/sample.txt
```

###   `append` to a file <br>

```python
f = open("data_file/sample.txt", "a")
```

```python
f.tell()
```

```python
f.writelines(["Another line was appended\n",
              "What will it look like now?\n", 
              "Lets check it out\n"])
f.close()
```

```python
f = open("data_file/sample.txt", "r")

f.readlines()
```


### <br>* `r+` method will overwrites the content
